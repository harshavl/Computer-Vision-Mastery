**Accessing and manipulting Pixles**

```python
#Get the color at (0,0) pixels
( b,g, r ) = image[0,0]  #reverse r,g,b

print(f"Pixels at (0,0) - Red: {r}, Green: {g} and Blue {b}")
```

```python
# replace pixles color (0,0) to red

image[0,0] = (0,0,255)
(b,g,r) = image[0,0]

print(f"Pixels at (0,0) - Red: {r}, Green: {g} and Blue {b}")
```

```python
# get the center of the image,

w = image.shape[1]
h = image.shape[0] 

(cX, cY) = ( w //2, h//2 )

#top-left corner

tl = image[0:cX, 0:cY ]

plt.imshow(tl,cmap='gray')
```

```python
tr = image[0:cY, cX:w]
br = image[cY:h, cX:w]
bl = image[cY:h, 0:cX]

plt.imshow( tr) #top right corner
```

```python
plt.imshow(br) # bottom-right corner
```

```python
plt.imshow(bl) #bottom-left corner
```

```python
#change the color of a region of pixels

image[0:cY, 0:cX] = (0, 255, 0)
plt.imshow(image)
```

### Drawing <br>
cv2.line <br>
cv2.rectangle <br>
cv2.circle <br>

```python
import numpy as np
```

```python
# draw a green line from the top-left corner of our canvas to the bottom-right
canvas = np.zeros((300, 300, 3), dtype="uint8")

green = (0, 255, 0)

cv2.line(canvas, (0, 0), (300, 300), green)

plt.imshow( canvas)
```

```python
# draw a green 50x50 pixel square, starting at 10x10 and ending at 60x60

cv2.rectangle(canvas, (10, 10), (60, 60), green)
plt.imshow(canvas)
```

```python
# draw a 3 pixel thick red line from the top-right corner to the bottom-left
red = (0, 0, 255)
cv2.line(canvas, (300, 0), (0, 300), red, 3)
plt.imshow(canvas)
```

```python
# draw another rectangle, this time we'll make it red and 5 pixels thick
cv2.rectangle(canvas, (50, 200), (200, 225), red, 5)
plt.imshow(canvas)
```

```python
blue = (255, 0, 0)
cv2.rectangle(canvas, (200, 50), (225, 125), blue, -1)
plt.imshow(canvas)
```

**Circles**

```python
# draw a white circle at the center of the canvas from 25 pixels to 150 pixels

canvas = np.zeros((300, 300, 3), dtype="uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

for r in range(0, 175, 25):
	cv2.circle(canvas, (centerX, centerY), r, white)
    
plt.imshow(canvas)
```

```python
# draw 25 random circles

for i in range(0,25):
    #randomly from 5 and 200
    radius = np.random.randint(5, high=200 )
    color = np.random.randint(0, high=256, size=(3,)).tolist()
    pt = np.random.randint(0, high=300, size=(2,))
    
    #draw circle random cicles
    cv2.circle( canvas, tuple(pt), radius, color, -1 )
    
plt.imshow(canvas)
```
