## Lighting and color spaces <br>
<br>

A color models: RGB, HSV, Lab and grayscale;<br>



```python
import numpy as np
import imutils
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file ="/mnt/c/cnn_dl/drraj.jpg"
image = cv2.imread(file)
```

```python
# RGB color

for (name,chan) in zip(("B","G","R"), cv2.split(image)):
    plt.figure()
    plt.imshow(chan)
    plt.show(name)
    
cv2.destroyAllWindows()
```

```python
# HSV: 
#Hue: Pure color; Satutration: How 'white' the color is; Value: control the lightness of color;

#convert image to the HSV color 
hsv = cv2.cvtColor( image, cv2.COLOR_BGR2HSV )
plt.imshow( hsv )
```

```python
#loop over each of the individual channels and display:

for (name,chan) in zip(('H','S','V'), cv2.split(hsv)):
    plt.figure()
    plt.imshow(chan)
```

```python
#L*a*b: 

lab = cv2.cvtColor( image, cv2.COLOR_BGR2LAB )
plt.imshow(lab)
```

```python
#each channels

for (name,chan) in zip( ("L*", "a*","b*"), cv2.split(lab)):
    plt.figure()
    plt.imshow(chan)
    
cv2.destroyAllWindows()
```

```python
#Grayscale: 

gray = cv2.cvtColor( image, cv2.COLOR_BGR2GRAY )
plt.imshow(gray)
```
