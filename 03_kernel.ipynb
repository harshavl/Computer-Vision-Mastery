**Kernels** or convolution matrix as a tiny matrix used for blurring, sharpening, edge detection, edge detection, Morphological operations and imgae processing. <br>
cv2.getStructuringElement() <br>


Morphological operations: Erosion, Dilation, Opening, Closing, Morphological gradient, Black hat, Top(white) hat. <br>


```python
import numpy as np
import imutils
import cv2

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file ="/mnt/c/cnn_dl/pyimagesearch_logo_noise.png"
image = cv2.imread(file)

plt.imshow( image )
```

**Erosion** useful in small blobs in an image or disconnecting two connected objects.


```python
gray = cv2.cvtColor( image, cv2.COLOR_BGR2GRAY )
plt.imshow( gray )
```

```python
#  inste

eroded = cv2.erode( gray.copy(), None, iterations=3 ) #instead on NOne, can supply own custom strcuturing element; by default 3x3 8-neighborhood
plt.imshow( eroded )
```

```python
eroded = cv2.erode( gray.copy(), None, iterations=10  )
plt.imshow( eroded) #Observation: check how and is disconneted from the circle;
```

**Dilation** will grow the foreground pixles.<br>
Usage: Joining broken parts of an image together.

```python
cv2.destroyAllWindows()

dilated = cv2.dilate(gray.copy(), None, iterations=0 )
plt.imshow( dilated)
```

```python
dilated = cv2.dilate(gray.copy(), None, iterations=3 )
plt.imshow( dilated) # background dark
```

**Opening** used to remove small blobs from an image; first, an erosion to remove the small blobs and then dilation is applied to regrow the size of the original object.

```python
cv2.destroyAllWindows()
plt.imshow(image)

kernelSizes = [(3, 3), (5, 5), (7, 7)]

for kernelsize in kernelSizes:
    kernel = cv2.getStructuringElement( cv2.MORPH_RECT, kernelsize)
    opening = cv2.morphologyEx( gray, cv2.MORPH_OPEN, kernel )
    plt.figure()
    plt.imshow( opening )
```

**Morphological Gradient** difference between the dilation and erosion.<br>
Usage: to determine the outline object of an image.

```python
cv2.destroyAllWindows()


for kernelsize in kernelSizes:
    kernel = cv2.getStructuringElement( cv2.MORPH_RECT, kernelsize )
    gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, kernel )
    plt.figure()
    plt.imshow(gradient)
```
